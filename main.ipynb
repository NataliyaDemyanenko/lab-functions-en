```python
def greatest(lst):
    max_element = 0
    for i in lst:
        if i > max_element:
            max_element = i
    return max_element

a=[1,2,3,4,5,6,7,7,8]
greatest(a)
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    factorial = 1
    for i in range (1,n+1):
        factorial *=i
    return factorial

a=3
factorial(a)
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

```python
def unique(lst_un):
    lst_output=[]
    for i in lst_un:
        if i not in lst_output:
            lst_output.append(i)
    return lst_output

        
a=[1,2,3,4,5,6,7,7,8]
unique(a)
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... ðŸ§`

```python
def mode_counter(arr):
    frequency={}
    for i in arr:
        if i in frequency:
            frequency[i]+=1
        else:
            frequency[i]=1

    max = 0
    mode = 0
    for i, count in frequency.items():
        if count > max:
            max = count
            mode = i
    
    return mode

a=[1,2,3,4,5,6,7,7,8]
mode_counter(a)
```

```python
# This will test your function 
test_mode(mode_counter)
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. ðŸ˜‰`

```python
def st_dev(list_sd):
    
    mean = sum(list_sd)/len(list_sd)
    sum_of_diff=0
    
    for i in list_sd:
        sum_of_diff += (i-mean)**2

    stdev = (sum_of_diff/(len(list_sd)-1))**(0.5)
    return stdev

a=[1,2,3,4,5,6,7,7,8]
st_dev(a)
```

```python
# This will test your function 
test_stdev(st_dev)
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
    alphabet = 'abcdefghijklmnopqrstuvwxyz'

    string = string.lower()
    
    letter_count = {letter: 0 for letter in alphabet}
    for i in alphabet:
        if i in string:
            letter_count[i]+=1 
        else:
            letter_count[i]=0

    for value in letter_count.values():
        if value == 0:
            return False
    
    return True

pangram('a,b,c,d,e,f,g,h,i,j,k,l,m,n,o,p,q,r,s,t,u,v,w,x,y,z,6')
```

```python
# This will test your function 
test_pangram(pangram)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ðŸ¤ª`

```python
def sort_alpha(string):
    list_of_words=string.split(",")
    sorted_list=[]
    sorted_list = sorted(list_of_words)
    new_string = ''
    for i in sorted_list:
        new_string +=i 
        new_string +=','
        
    new_string = new_string[:-1]
    return new_string
    

a='hello,how,are,you'
sort_alpha(a)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
    if len(password) < 8:
        return False
      
    has_lower = False
    has_upper = False
    has_digit = False
    has_special = False

    for char in password:
        if char.islower():
            has_lower = True
        elif char.isupper():
            has_upper = True
        elif char.isdigit():
            has_digit = True
        elif char in '!@#$%^&*()-+':
            has_special = True

    if has_lower == True and has_upper == True and has_digit == True and has_special == True:
        return True
    else:
        return False

password_check='12345Yy0'
check_pass(password_check)
```

```python
# This will test your function 
test_pass(check_pass)
```
